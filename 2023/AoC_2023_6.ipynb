# Day 6 (2023-12-06), Exercise a

```python
with open("input_6.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
import re
race_times = [int(t) for t in re.findall(r'\d+', lines[0])]
race_distance_records = [int(d) for d in re.findall(r'\d+', lines[1])]
races = []
for i in range(len(race_times)):
    races.append({"time": race_times[i], "distance_record": race_distance_records[i]})
races
```

```python
win_counts = []
for r in races:
    win_count = 0
    for speed in range(r['time']+1):
        distance = speed * (r['time']-speed)
        if distance > r['distance_record']:
            win_count += 1
    win_counts.append(win_count)
win_counts
```

```python
from math import prod
prod(win_counts)
```

# Day 6 (2023-12-06), Exercise b

```python
race_time = int(''.join([str(t) for t in race_times]))
distance_record = int(''.join([str(d) for d in race_distance_records]))
distance_record
```

```python
# Method using ABC-formula (s=speed, T=race_time, D=distance_record, t=time boat is moving)
# s*t = (T-t)*t > D
# (T-t)*t - D > 0
# -t^2 + Tt - D > 0
# ABC-formula: x = (-b +- sqrt(b^2-4ac))/(2a)
# All t's between min and max t-solution are valid solutions to break the record.

from math import sqrt
t_first = (race_time + sqrt(race_time*race_time - 4*distance_record))/2
t_second = (race_time - sqrt(race_time*race_time - 4*distance_record))/2
print(t_first, t_second)
```

```python
num_records_broken = int(max(t_first, t_second)) - int(min(t_first, t_second))
num_records_broken
```
