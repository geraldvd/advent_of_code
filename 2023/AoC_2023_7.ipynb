# Day 7 (2023-12-07), Exercise a

```python
with open("input_7.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
from operator import itemgetter
```

```python
def get_type(hand):
    # Types: 7=5 of a kind, 6=4 of a kind, 5=full house, 4=three of a kind, 3=two pair, 2=pair, 1=high card
    card_hist = {}
    for c in hand:
        card_hist[c] = card_hist[c] + 1 if c in card_hist.keys() else 1
    counts = list(card_hist.values())
    if max(counts) == 5: return 7
    if max(counts) == 4: return 6
    if 3 in counts and 2 in counts: return 5
    if 3 in counts: return 4
    if 2 in counts and 1 in counts and len(counts) == 3: return 3
    if 2 in counts: return 2
    return 1


def hand_to_card_value_list(hand):
    # Using letters for values, since that fits in 1 character
    card_values = {'A': 'Z', 'K': 'Y', 'Q': 'X', 'J': 'W', 'T': 'V', '9': 'U', '8': 'T', '7': 'S', '6': 'R', '5': 'Q', '4': 'P', '3': 'O', '2': 'N',}
    return ''.join([card_values[h] for h in hand])
```

```python
# Determine types for each hand, and store in dict
hands_by_type = {}
for l in lines:
    hand, bet = l.split(' ')
    t = get_type(hand)
    if t in hands_by_type.keys():
        hands_by_type[t].append((hand, bet))
    else:
        hands_by_type[t] = [(hand, bet)]

# Go through each type from weakest to best hand
total_bet = 0
rank = 1
for t in range(1, 8):
    hands_dict = {hand_to_card_value_list(h[0]):h for h in hands_by_type[t]}
    sorted_keys = sorted(hands_dict, key=itemgetter(0, 1, 2, 3, 4))
    for sk in sorted_keys:
        total_bet += int(hands_dict[sk][1]) * rank
        rank += 1
print(total_bet)
```

# Day 7 (2023-12-07), Exercise b

```python
import re
```

```python
card_values_no_joker = ['A', 'K', 'Q', 'T', '9', '8', '7', '6', '5', '4', '3', '2']
```

```python
def get_all_possible_hands(hand):
    joker_idx = [c.start() for c in re.finditer('J', hand)]
    hands = [hand]
    while len(joker_idx):
        ji = joker_idx.pop(0)
        new_hands = []
        for h in hands:
            for k in card_values_no_joker:
                new_hands.append(h[:ji] + k + h[ji + 1:])
        hands = new_hands
    return hands

def get_best_type(hand):
    hands = get_all_possible_hands(hand)
    best_hand_type = 0
    for h in hands:
        assert 'J' not in h, "There shouldn't be a joker in the hand."
        t = get_type(h)
        if t > best_hand_type:
            best_hand_type = t
    return best_hand_type

def hand_to_card_value_list(hand):
    # Using letters for values, since that fits in 1 character. Note: Joker demoted to lowest rank.
    card_values = {'A': 'Z', 'K': 'Y', 'Q': 'X', 'T': 'V', '9': 'U', '8': 'T', '7': 'S', '6': 'R', '5': 'Q', '4': 'P', '3': 'O', '2': 'N', 'J': 'M'}
    return ''.join([card_values[h] for h in hand])
```

```python
# Determine types for each hand, and store in dict
hands_by_type = {}
for l in lines:
    hand, bet = l.split(' ')
    t = get_best_type(hand)
    if t in hands_by_type.keys():
        hands_by_type[t].append((hand, bet))
    else:
        hands_by_type[t] = [(hand, bet)]
```

```python
# Go through each type from weakest to best hand
total_bet = 0
rank = 1
for t in range(1, 8):
    hands_dict = {hand_to_card_value_list(h[0]):h for h in hands_by_type[t]}
    sorted_keys = sorted(hands_dict, key=itemgetter(0, 1, 2, 3, 4))
    for sk in sorted_keys:
        total_bet += int(hands_dict[sk][1]) * rank
        rank += 1
print(total_bet)
```
