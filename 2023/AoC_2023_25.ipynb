```python
with open("input_25_dummy.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
import numpy as np
```

```python
def initialize_pairs(lines):
    pairs = []
    for l in lines:
        key, value = l.split(': ')
        values = value.split(' ')
        for v in values:
            pairs.append(tuple(sorted([key, v])))
    return list(set(pairs))

def get_all_keys(pairs):
    all_keys = []
    for p in pairs:
        all_keys += [p[0], p[1]]
    return all_keys

def get_pair_hash(pairs):
    pair_hash = {}
    for p in pairs:
        if p[0] not in pair_hash.keys():
            pair_hash[p[0]] = []
        pair_hash[p[0]].append(p[1])
        if p[1] not in pair_hash.keys():
            pair_hash[p[1]] = []
        pair_hash[p[1]].append(p[0])
    return pair_hash

def count_pair_groups(pairs):
    pair_hash = get_pair_hash(pairs)
    component_group = []
    p = pairs[0]
    component_group += list(p)

def count_key_groups(keys, pair_hash):
    first_key = list(pair_hash.keys())[0]
    key_groups = {}
    key_groups[first_key] = set()
    len_set = 0
    for k in pair_hash.keys():
        for k2 in pair_hash[k]:
            key_groups[first_key].add(k2)
    return key_groups


pairs = initialize_pairs(lines)
all_keys = get_all_keys(pairs)
pair_hash = get_pair_hash(pairs)
```
