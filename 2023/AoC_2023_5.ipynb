# Day 5 (2023-12-05), Exercise a

```python
with open("input_5.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
map_sequence = {"seed-to-soil": 1, "soil-to-fertilizer": 2, "fertilizer-to-water": 3, "water-to-light": 4, "light-to-temperature": 5, "temperature-to-humidity": 6, "humidity-to-location": 7}
```

```python
seeds = lines[0].removeprefix("seeds: ").split(" ")

# Split lines into different corresponding maps
map_headers_idx = [i for i, l in enumerate(lines) if "map" in l]
maps = {}
for i, i_map in enumerate(map_headers_idx):
    if i_map == map_headers_idx[len(map_headers_idx)-1]:
        maps[lines[i_map].removesuffix("map:").strip()] = lines[i_map+1:] 
    else:
        maps[lines[i_map].removesuffix("map:").strip()] = lines[i_map+1:map_headers_idx[i+1]-1]
```

```python
def seed2location(seed):
    # Walk through each map, where s is first seed, then soil, then fertilizer, etc.
    s = int(seed)
    for m in maps.values():
        for l in m:
            ids = [int(i) for i in l.split(" ")]
            # Check whether s is in source range; if not: no action required, since s does not change for next step.
            if ids[1] <= s < ids[1]+ids[2]:
                s = ids[0] + (s-ids[1])
                break
    return s

def location2seed(location):
    # Walk through each map in reverse
    s = int(location)
    for m in reversed(maps.values()):
        for l in m:
            ids = [int(i) for i in l.split(" ")]
            # Check whether s is in source range; if not: no action required, since s does not change for next step.
            if ids[0] <= s < ids[0]+ids[2]:
                s = ids[1] + (s-ids[0])
                break
    return s
```

```python
min_location = -1
for seed in seeds:
    s = seed2location(seed)
    assert s != min_location, "Multiple minimal locations found"
    if s < min_location or min_location < 0:
        min_location = s
    print(f"seed {seed} has location {s}")
print(f"minimum location: {min_location}")
```
