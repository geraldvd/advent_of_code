# Day 11 (2023-12-11), Exercise a

```python
with open("input_11.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
lines_transpose = list(map(list, zip(*lines)))
```

```python
double_count_rows = [i for i, l in enumerate(lines) if '#' not in l]
double_count_columns = [i for i, l in enumerate(lines_transpose) if '#' not in l]
```

```python
galaxies = []
for y, l in enumerate(lines):
    for x, c in enumerate(l):
        if c == '#':
            galaxies.append((x, y))
```

```python
total_distance = 0
for i in range(len(galaxies)):
    for j in range(i+1, len(galaxies)):
        g1, g2 = galaxies[i], galaxies[j]
        dcr = sum([1 for i in double_count_rows if min(g1[1], g2[1]) < i < max(g1[1], g2[1])])
        dcc = sum([1 for j in double_count_columns if min(g1[0], g2[0]) < j < max(g1[0], g2[0])])
        distance = (max(g1[1], g2[1]) - min(g1[1], g2[1]) + dcr) + (max(g1[0], g2[0]) - min(g1[0], g2[0]) + dcc)
        total_distance += distance
total_distance
```
