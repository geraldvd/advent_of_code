# Day 9 (2023-12-09), Exercise a

```python
with open("input_9.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
def diffs(seq):
    # Return sequence of differences (len will be 1 shorter than original)
    return [seq[i+1]-s for i, s in enumerate(seq[:-1])]

def iterate_seq(seq):
    # Build list of differences
    sub_seqs = []
    while len(set(seq)) > 1:
        sub_seqs.append(seq)
        seq = diffs(seq)
    sub_seqs.append(seq)
    
    # Iterate back through list to find next history value
    next_value = 0
    while len(sub_seqs):
        ss = sub_seqs.pop(-1)
        next_value = ss[-1]+next_value
    return next_value
```

```python
# Go through all original sequences in input file and add next values
sum_next_values = 0
for l in lines:
    seq = [int(n) for n in l.split(' ')]
    sum_next_values += iterate_seq(seq)
sum_next_values
```

```python
# Identical approach as part a, but just reverse order of original list
sum_next_values = 0
for l in lines:
    seq = [int(n) for n in l.split(' ')][::-1]
    sum_next_values += iterate_seq(seq)
sum_next_values
```
