```python
with open("input_16.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
grid = [l.strip() for l in lines]
assert len(grid) == len(grid[0]), "Assuming grid is square"
```

```python
def single_move(position, direction):
    # First determine new position based on previous position and direction
    match direction:
        case 'R': new_pos = (position[0], position[1]+1)
        case 'L': new_pos = (position[0], position[1]-1)
        case 'U': new_pos = (position[0]-1, position[1])
        case 'D': new_pos = (position[0]+1, position[1])
    # Move within grid
    if 0 <= new_pos[0] < len(grid) and 0 <= new_pos[1] < len(grid):
        # Determine new direction (can be multiple)
        match grid[new_pos[0]][new_pos[1]]:
            case '/':
                match direction:
                    case 'R': new_dir = 'U'
                    case 'L': new_dir = 'D'
                    case 'U': new_dir = 'R'
                    case 'D': new_dir = 'L'
            case '\\':
                match direction:
                    case 'R': new_dir = 'D'
                    case 'L': new_dir = 'U'
                    case 'U': new_dir = 'L'
                    case 'D': new_dir = 'R'
            case '-':
                match direction:
                    case 'U': new_dir = ['L', 'R']
                    case 'D': new_dir = ['L', 'R']
                    case _: new_dir = direction
            case '|':
                match direction:
                    case 'L': new_dir = ['U', 'D']
                    case 'R': new_dir = ['U', 'D']
                    case _: new_dir = direction
            case _: 
                new_dir = direction
        
        # Return all (new) beams
        if type(new_dir) == str:
            return [(new_pos, new_dir)]
        new_beams = []
        for nd in new_dir:
            new_beams.append((new_pos, nd))
        return new_beams
    # Move goes outside grid
    else:
        return []
```

```python
# Each beam is tupple with: tuple (i,j) and string (direction: L/R/U/D)
def find_energized(start_beam):
    beams = [start_beam]
    encountered_beams = set()
    energized = set()

    counter = 0

    while len(beams) and counter < 10000000:
        position, direction = beams.pop(0)
        new_beams = [nb for nb in single_move(position, direction) if nb not in encountered_beams]
        for nb in new_beams:
            encountered_beams.add(nb)
            energized.add(nb[0])
        beams += new_beams
        # if counter % 1000:
        #     print(beams)
        counter += 1
    return len(energized)

start_beam = ((0,-1), 'R') # start just left of the grid
find_energized(start_beam)
```

```python
energized_configs = []
for r in range(0, len(grid)):
    for c in range(0, len(grid)):
        if r != 0 and r != len(grid)-1 and c != 0 and c != len(grid)-1:
            continue
        position = (r, c)
        directions = []
        if r == 0: directions.append('D')
        if r == len(grid)-1: directions.append('U')
        if c == 0: directions.append('R')
        if c == len(grid)-1: directions.append('L')
        for direction in directions:
            match direction:
                case 'R': start_beam = ((r, c-1), direction)
                case 'L': start_beam = ((r, c+1), direction)
                case 'U': start_beam = ((r+1, c), direction)
                case 'D': start_beam = ((r-1, c), direction)
            
            energized_configs.append(find_energized(start_beam))
max(energized_configs)
```
