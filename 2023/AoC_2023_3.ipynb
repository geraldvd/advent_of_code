# Day 3 (2023-12-03), Exercise a

```python
with open("input_3.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
# Get all numbers and coordinates of first digit
numbers_found = []
number_found = ''
for y, l in enumerate(lines):
    # Check if a number was found at the end of previous line
    if number_found != '':
        numbers_found.append({"num": number_found, "coord": (len(lines[0])-len(number_found)-1, y-1)})
        number_found = ''
    # Scan through line for numbers
    for x, s in enumerate(l):
        if not s.isnumeric():
            if number_found != '':
                numbers_found.append({"num": number_found, "coord": (x-len(number_found), y)})
                number_found = ''
        else:
            number_found += s
        
# Check whether number is surrounded by symbol, and if so, add to total
total = 0
for nf in numbers_found:
    is_engine_part = 0
    for x in range(nf['coord'][0]-1, nf['coord'][0]+len(nf['num'])+1):
        for y in range(nf['coord'][1]-1, nf['coord'][1]+2):
            if len(lines[0]) > x >= 0 and len(lines) > y >= 0 and not lines[y][x].isnumeric() and lines[y][x] != '.':
                is_engine_part += 1
    if is_engine_part:
        assert is_engine_part == 1, "Count of symbols > 1 found"
        total += int(nf['num'])

total
```

# Day 3 (2023-12-03), Exercise b

```python
from math import prod
```

```python
# Find all gear symbols
gears = {}
for y, l in enumerate(lines):
    for x, s in enumerate(l):
        if s == "*":
            gears[(x, y)] = {"coord": (x, y), "numbers_adjacent": []}

# List numbers adjacent to each gear symbol
for nf in numbers_found:
    for x in range(nf['coord'][0]-1, nf['coord'][0]+len(nf['num'])+1):
        for y in range(nf['coord'][1]-1, nf['coord'][1]+2):
            if len(lines[0]) > x >= 0 and len(lines) > y >= 0 and lines[y][x] == '*':
                gears[(x,y)]['numbers_adjacent'].append(int(nf['num']))

# Take sum-product of gears with exactly 2 adjacent numbers
valid_gear_ratios = [prod(g['numbers_adjacent']) for g in gears.values() if len(g['numbers_adjacent']) == 2]
sum(valid_gear_ratios)
```
