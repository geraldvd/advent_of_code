# Day 12 (2023-12-12), Exercise a

```python
with open("input_12_dummy.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
def row_is_valid(pattern, num_pounds):
    return [len(p) for p in pattern.split('.') if len(p)] == list(num_pounds)
```

```python
from itertools import combinations
```

```python
total_arrangements = 0
for l in lines:
    # Split string
    pattern, pound_count = l.split(" ")
    pound_count = [int(n) for n in pound_count.split(",")] * 2
    pattern += '?' + pattern
    
    # Go through all combinations of pound/dot replacement for question marks
    num_pounds_left = sum(pound_count) - sum([1 for c in pattern if c == '#'])
    qmark_idx = [i for i, c in enumerate(pattern) if c == '?']
    arrangements = 0
    for c in combinations(qmark_idx, num_pounds_left):
        s = pattern[:c[0]]
        for i, qi in enumerate(c):
            if i == len(c) - 1:
                s += '#' + pattern[qi+1:]
            else:
                s += '#' + pattern[qi+1:c[i+1]]
        s = s.replace('?', '.')
        arrangements += 1 if row_is_valid(s, pound_count) else 0
    print(arrangements)
    total_arrangements += arrangements
total_arrangements
```
