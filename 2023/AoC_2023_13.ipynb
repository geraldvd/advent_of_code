# Day 13 (2023-12-13), Exercise a

```python
with open("input_13.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
def find_symmetry_horizontal(pattern):
    for split_idx in range(1, len(pattern)):
        top, bottom = pattern[:split_idx][::-1], pattern[split_idx:]
        symmetric = True
        for i in range(min(len(top), len(bottom))):
            if top[i] != bottom[i]:
                symmetric = False
                break
        if symmetric:
            return len(top)
    return -1
```

```python
patterns = [l.split('\n') for l in '\n'.join(lines).split('\n\n')]
total_symmetry_count = 0
for p in patterns:
    # Check horizontal first
    symmetry_count = find_symmetry_horizontal(p)
    # If not found: transpose to check vertical 
    if symmetry_count < 0:
        p_transpose = list(map(list, zip(*p)))
        symmetry_count = find_symmetry_horizontal(p_transpose)
    else:
        symmetry_count *= 100
    total_symmetry_count += symmetry_count
total_symmetry_count
```

```python
def find_single_smudge_mirror(pattern):
    # Convert # and . into 1 and 0
    p = [pi.replace('#', '1').replace('.', '0') for pi in pattern]
    
    # Find mirror line by checking where there is an absolute difference within a pattern of exactly 1
    mirror_line = -1
    for m in range(1, len(p)):
        if m <= int(len(p)/2):
            p_test = p[:2*m]
        else:
            p_test = p[m-(len(p)-m):]
        p1, p2 = p_test[:int(len(p_test)/2)], p_test[int(len(p_test)/2):][::-1]
        
        diff = sum([sum([abs(int(p2[i][j])-int(p1[i][j])) for j in range(len(p1[i]))]) for i in range(len(p1))])
        if diff == 1:
            mirror_line = m
    return mirror_line
```

```python
patterns = [l.split('\n') for l in '\n'.join(lines).split('\n\n')]
total_symmetry_count = 0
for p in patterns:
    # Check horizontal first
    symmetry_count = find_single_smudge_mirror(p)
    # If not found: transpose to check vertical 
    if symmetry_count < 0:
        p_transpose = [''.join(l) for l in list(map(list, zip(*p)))]
        symmetry_count = find_single_smudge_mirror(p_transpose)
    else:
        symmetry_count *= 100
    total_symmetry_count += symmetry_count
total_symmetry_count
```
