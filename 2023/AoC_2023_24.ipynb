```python
with open("input_24.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
import numpy as np
```

```python
def test_intersection(px1, vx1, px2, vx2, py1, vy1, py2, vy2):
    divisor1 = vx2 - vx1
    divisor2 = vy2 - vy1
    if divisor1 == 0 or divisor2 == 0:
        # Divide by zero, parallel lines 
        return None, None, None, None, None, None, None, None # assuming: t<0 is invalid
    t1 = float(px1 - px2) / divisor1
    t2 = float(py1 - py2) / divisor2
    # t = ((px1-px2) - (py1-py2))/((vy1-vy2) - (vx1-vx2))
    x1 = px1 + vx1*t1
    y1 = py1 + vy1*t1
    x2 = px2 + vx2*t2
    y2 = py2 + vy2*t2
    # print(t1, t2, x1, x2, y1, y2)
    return t1, t2, x1, y1, x2, y2, divisor1, divisor2
```

```python
# Initialize position and velocity arrays
raw_numbers = [[tuple([int(i) for i in n.split(', ')]) for n in l.split(' @ ')] for l in lines]
p = np.zeros((len(raw_numbers), 3))
v = np.zeros((len(raw_numbers), 3))
for i, rn in enumerate(raw_numbers):
    p[i] = rn[0]
    v[i] = rn[1]

p_xy = p[:, :2]
v_xy = v[:, :2]
```

```python
# Calculate position for upper and lower bounds
p_lower = 200000000000000
p_upper = 400000000000000
# p_lower = 7
# p_upper = 27

# Intersect all pairs
valid_intersections = 0
for i in range(len(p_xy)):
    p1 = p_xy[i]
    v1 = v_xy[i]
    for j in range(i+1, len(p_xy)):
        p2 = p_xy[j]
        v2 = v_xy[j]
        t1, t2, x1, y1, x2, y2, divisor1, divisor2 = test_intersection(p1[0], v1[0], p2[0], v2[0], p1[1], v1[1], p2[1], v2[1])
        if t1 is not None and t2 is not None and t1 > 0 and t2 > 0 and p_lower <= x1 <= p_upper and p_lower <= y1 <= p_upper and p_lower <= x2 <= p_upper and p_lower <= y2 <= p_upper:
            # print(p1, p2)
            valid_intersections += 1
valid_intersections
# 21064 too high
# 13341 too low
```
