# Day 2 (2023-12-02), Exercise a

```python
with open("input_2.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
check_bag = {"red": 12, "green": 13, "blue": 14}
```

```python
lines2 = [l.lstrip("Game").strip() for l in lines]
games = dict([l.split(':') for l in lines2])
for k, v in games.items():
    v = v.split(";")
    v = [vi.strip().split(", ") for vi in v]
    games[k] = [dict([vii.split(" ")[::-1] for vii in vi]) for vi in v]
```

```python
sum_valid_games = 0
for game_num, game_sets in games.items():
    game_valid = True
    for s in game_sets:
        set_valid = True
        for color, num in s.items():
            if int(num) > check_bag[color]:
                set_valid = False
                break
        if not set_valid:
            game_valid = False
            break
    if game_valid:
        sum_valid_games += int(game_num)
        
print(sum_valid_games)
```

# Day 2 (2023-12-02), Exercise b

```python
from math import prod
```

```python
sum_of_powers = 0
for game_num, game_sets in games.items():
    min_nums = {"red": 0, "green": 0, "blue": 0}
    for s in game_sets:
        for color, num in s.items():
            if int(num) > min_nums[color]:
                min_nums[color] = int(num)
    sum_of_powers += prod(min_nums.values())
        
print(sum_of_powers)
```
