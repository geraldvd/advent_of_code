# Day 8 (2023-12-08), Exercise a

```python
with open("input_8.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
left_right = lines[0]
left_right = [int(lr) for lr in left_right.replace('L', '0').replace('R', '1')]
```

```python
network_map = {}
for l in lines[2:]:
    cur_loc = l[:3]
    left = l[7:10]
    right = l[12:15]
    network_map[cur_loc] = (left, right)
```

```python
cur_loc = "AAA"
lr_count = 0
lr_index = 0
while cur_loc != "ZZZ":
    cur_loc = network_map[cur_loc][left_right[lr_index]]
    lr_index += 1
    lr_count += 1
    if lr_index == len(left_right):
        lr_index = 0
print(lr_count)
```

```python
start_nodes = [n for n in network_map.keys() if n[-1] == "A"]
end_nodes = [n for n in network_map.keys() if n[-1] == "Z"]
start_nodes, end_nodes
```

```python
# Find pattern from each start node to any end node
# Conclusion: always a "round" LR sequence, and each **A ends at a different **Z
for n in range(len(start_nodes)):
    cur_loc = start_nodes[n]
    lr_count = 0
    lr_index = 0
    while cur_loc[-1] != "Z":
        cur_loc = network_map[cur_loc][left_right[lr_index]]
        lr_index += 1
        lr_count += 1
        if lr_index == len(left_right):
            lr_index = 0
    print(f"{start_nodes[n]} --> {cur_loc}: {lr_count} steps ({lr_count/len(left_right)} left right sequences)")
```

```python
# Find pattern from each end node to any end node
# Conclusion: starting at an end node always end at the same end node, AND uses the same number of steps as from **A to that **Z
lr_sequences = []
steps = []
for n in range(len(end_nodes)):
    cur_loc = end_nodes[n]
    lr_count = 0
    lr_index = 0
    while cur_loc[-1] != "Z" or not lr_count:
        cur_loc = network_map[cur_loc][left_right[lr_index]]
        lr_index += 1
        lr_count += 1
        if lr_index == len(left_right):
            lr_index = 0
    lr_sequences.append(lr_count/len(left_right))
    steps.append(lr_count)
    print(f"{end_nodes[n]} --> {cur_loc}: {lr_count} steps ({lr_count/len(left_right)} left right sequences)")
```

```python
# Answer is the product of all full left_right sequences, times the length of the sequence
from math import prod
int(prod(lr_sequences))*len(left_right)
```
