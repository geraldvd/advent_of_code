```python
with open("input_1.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
lines2 = [[s for s in l if s.isnumeric()] for l in lines]
lines_sum = [int(l[0]+l[-1]) for l in lines2]
sum(lines_sum)
```

# Day 1 (2023-12-01), Exercise b

```python
number_strings = {"one": "1", "two": "2", "three": "3", "four": "4", "five": "5", "six": "6", "seven": "7", "eight": "8", "nine": "9"}
```

```python
lines3 = [l for l in lines]
for ns in number_strings.items():
    # Based on hint from Reddit, number can share first and last letter (e.g., eighthree).
    # To solve this, a number string is replaced with number, prefixed and suffixed with first and last letter respectively.
    # E.g., one => o1e, eight => e8t, etc. 
    lines3 = [l.replace(ns[0], ns[0][0]+ns[1]+ns[0][-1]) for l in lines3]
lines4 = [[s for s in l if s.isnumeric()] for l in lines3]
lines_sum2 = [int(l[0]+l[-1]) for l in lines4]
sum(lines_sum2)
```
