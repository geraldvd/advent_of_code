# Day 4 (2023-12-04), Exercise a

```python
with open("input_4.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
lines2 = [l.split(":")[1].strip() for l in lines]
lines3 = [l.split(" | ") for l in lines2]
winning_numbers = [l[0].split(" ") for l in lines3]
all_numbers = [l[1].split(" ") for l in lines3]

# Remove blanks and convert to int
winning_numbers = [[int(n.strip()) for n in wn if n.strip()] for wn in winning_numbers]
all_numbers = [[int(n.strip()) for n in an if n.strip()] for an in all_numbers]
```

```python
score = 0
for i in range(len(winning_numbers)):
    score_i = 0
    for wn in winning_numbers[i]:
        if wn in all_numbers[i]:
            # First match gets a score of 1, every next match doubles score
            score_i = 2*score_i if score_i else 1
    score += score_i
```

# Day 4 (2023-12-04), Exercise b

```python
original_scratchcards = {}
for i in range(len(winning_numbers)):
    original_scratchcards[i+1] = (winning_numbers[i], all_numbers[i])
```

```python
# Counter of each scratchcard (initialized with 1 for original, and copies added below)
scratchcards = {i:1 for i in original_scratchcards.keys()}
last_scratchcard_idx = max(scratchcards.keys())
```

```python
for i in scratchcards.keys():
    # Compute number of matching winning numbers
    num_matches = 0
    for wn in original_scratchcards[i][0]:
        if wn in original_scratchcards[i][1]:
            num_matches += 1
    # Add new cards based on number of matches
    if num_matches:
        for j in range(i+1, min(last_scratchcard_idx+1, i+num_matches+1)):
            # Add new cards for each card i (not exceeding last index)
            scratchcards[j] += scratchcards[i]
```
