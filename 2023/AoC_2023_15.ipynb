# Day 15 (2023-12-15), Exercise a

```python
with open("input_15.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
def hash_alg(seq):
    current_value = 0
    for s in seq:
        current_value = (17 * (current_value + ord(s))) % 256
    return current_value
```

```python
sequence = lines[0].split(',')
total_value = sum([hash_alg(seq) for seq in sequence])
total_value
```

# Day 15 (2023-12-15), Exercise b

```python
lenses = {i:[] for i in range(256)}
focal_lengths = {i:{} for i in range(256)}
for seq in sequence:
    label = seq.removesuffix('-').split('=')
    focal_length = label[1] if len(label)>1 else None
    label = label[0]
    box = hash_alg(label)
    
    if focal_length:
        focal_lengths[box][label] = focal_length
        if label not in lenses[box]:
            lenses[box].append(label)
    else:
        focal_lengths[box][label] = 0
        if label in lenses[box]:
            lenses[box].remove(label)
            
focussing_power = 0
for i in range(256):
    for j, l in enumerate(lenses[i]):
        focussing_power += (1+i) * (1+j) * int(focal_lengths[i][l])
focussing_power
```
