# Day 14 (2023-12-14), Exercise a

```python
with open("input_14.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
lines_transpose = list(map(list, zip(*lines)))
```

```python
def rotate_left(dish):
    return [''.join(l) for l in list(map(list, zip(*dish)))[::-1]]

def rotate_right(dish):
    return [''.join(l) for l in list(map(list, zip(*dish[::-1])))]

def tilt_west(dish):
    dish_out = []
    for l in dish:
        l2 = []
        # Find all solid rocks (#)
        solid_rocks = [i for i, c in enumerate(l) if c == '#']
        # Roll all rolling rocks to the left before first solid rock
        if not len(solid_rocks):
            rolling_rocks_count = sum([1 for r in l if r == 'O'])
            l2 += ['O']*rolling_rocks_count + ['.']*(len(l)-rolling_rocks_count)
        else:
            rolling_rocks_count = sum([1 for r in l[0:solid_rocks[0]] if r == 'O'])
            l2 += ['O']*rolling_rocks_count + ['.']*(len(l[0:solid_rocks[0]])-rolling_rocks_count)
        for i, sri in enumerate(solid_rocks):
            # Roll all rolling rocks against previous solid rock
            next_sri = solid_rocks[i+1] if i+1 < len(solid_rocks) else len(l)
            rolling_rocks_count = sum([1 for r in l[sri:next_sri] if r == 'O'])
            l2 += ['#'] + ['O']*rolling_rocks_count + ['.']*(len(l[sri:next_sri])-rolling_rocks_count-1)
        dish_out.append(l2)
    return dish_out

def calculate_load_north(dish):
    total_load = 0
    for i, l in enumerate(dish[::-1]):
        rolling_rocks_count = sum([1 for r in l if r == 'O'])
        total_load += rolling_rocks_count * (i+1)
    return total_load
```

```python
calculate_load_north(rotate_right(tilt_west(rotate_left(lines))))
```

```python
def tilt_north(dish):
    return rotate_right(tilt_west(rotate_left(dish)))

def cycle(dish):
    # rotate from North to East
    dish_N = tilt_north(dish)
    dish_W = tilt_north(rotate_right(dish_N))
    dish_S = tilt_north(rotate_right(dish_W))
    dish_E = tilt_north(rotate_right(dish_S))
    return rotate_right(dish_E)
```

```python
# First pattern occurence found
dish = lines.copy()
north_load = 0
i = 0
j = -1
search_offset = -1
while search_offset:
    j += 1
    # Condition i<200 used to make sure pattern stabilized
    while north_load != pattern_found[j] or i < 200:
        dish = cycle(dish)
        north_load = calculate_load_north(dish)
        i += 1
    # Correct answer is found when no remainder is found.
    search_offset = (1000000000-i)%len(pattern_found)

pattern_found[j]
```
