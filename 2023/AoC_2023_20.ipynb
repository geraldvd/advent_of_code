# Day 20 (2023-12-20), Exercise a

```python
with open("input_20.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
def initialize():
    # Prep input
    modules = {}
    conj_mod_keys = []
    for l in lines:
        k, v = l.split(' -> ')
        module_type = k[0] if k != 'broadcaster' else 'broadcaster'
        if module_type == '%':
            modules[k.removeprefix('%')] = [module_type, v.split(', '), False] # Last element is the state: Flip Flop is initially OFF
        elif module_type == '&':
            modules[k.removeprefix('&')] = [module_type, v.split(', '), {}] # Last element is the state: conj remembers list for each input (initially OFF)
            conj_mod_keys.append(k.removeprefix('&'))
        else:
            modules[k] = [module_type, v.split(', '), None] # Last element is the state: broadcaster has no state

    # Initialize conjunction states by finding number of inputs
    for conj_key in conj_mod_keys:
        for m in modules.keys():
            if conj_key in modules[m][1]:
                modules[conj_key][2][m] = False
    return modules

def flipflop(pulse, state):
    if pulse: return None
    else: return not state

def conjunction(pulse, state, source_mod_key):
    state[source_mod_key] = pulse
    return len(state.values()) != sum(state.values()), state

def button_press(modules):
    low_pulses = 1
    high_pulses = 0
    rx_value = None
    pulse_list = [(m, False, 'broadcaster') for m in modules['broadcaster'][1]] # First pulse is low, and source is broadcaster
    while len(pulse_list):
        mod_key, pulse_value, source_mod_key = pulse_list.pop(0)
        if pulse_value is True: high_pulses += 1
        elif pulse_value is False: low_pulses += 1
        if mod_key not in modules.keys():
            rx_value = pulse_value
            continue
        mod_type, rec_list, mod_state = modules[mod_key]
        if mod_type == '%':
            output = flipflop(pulse_value, mod_state)
            if output is not None:
                modules[mod_key][2] = output
                pulse_list += [(m, output, mod_key) for m in modules[mod_key][1]]
        if mod_type == '&':
            output, state = conjunction(pulse_value, mod_state, source_mod_key)
            modules[mod_key][2] = state
            pulse_list += [(m, output, mod_key) for m in modules[mod_key][1]]
    return low_pulses, high_pulses, modules, rx_value
```

```python
modules = initialize()
total_low = 0
total_high = 0

for b in range(1000):
    low_pulses, high_pulses, modules, _ = button_press(modules)
    total_high += high_pulses
    total_low += low_pulses
(total_low, total_high, total_low*total_high)
```
