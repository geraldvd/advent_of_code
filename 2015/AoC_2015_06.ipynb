```python
import re
import time
```

```python
with open("input_06.txt", "r") as fh:
    content = fh.readlines()

instructions = []
for c in content:
    result = re.findall(
        r"[turn on|toggle|turn off]\s+(\d+),(\d+)\s+through\s+(\d+),(\d+)", c
    )
    method = ""
    if "turn on" in c:
        method = "turn on"
    elif "turn off" in c:
        method = "turn off"
    elif "toggle" in c:
        method = "toggle"
    if len(result):
        nums = [int(r) for r in result[0]]
        instructions.append((method, (nums[0], nums[1]), (nums[2], nums[3])))
```

```python
# 6a
t0 = time.time()
lights = [[0 for _ in range(1000)] for _ in range(1000)]
for i in instructions:
    method, (x_min, y_min), (x_max, y_max) = i
    for y, l in enumerate(lights[y_min:y_max]):
        for x, _ in enumerate(l[x_min:x_max]):
            if method == "toggle":
                lights[y][x] = abs(lights[y][x] - 1)
            else:
                lights[y][x] = 1 if method == "turn on" else 0
print("6a:", sum([sum(l) for l in lights]), f"in {time.time() - t0} sec.")
# 410351 is too high
```
