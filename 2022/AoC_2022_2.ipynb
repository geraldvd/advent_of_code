```python
with open("input_2.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
hand_values = {
    'A': 1,
    'B': 2,
    'C': 3,
    'X': 1,
    'Y': 2,
    'Z': 3,
}
win_map = {
    ('A', 'X'): 'DRAW',
    ('B', 'X'): 'OPPONENT',
    ('C', 'X'): 'ME',
    ('A', 'Y'): 'ME',
    ('B', 'Y'): 'DRAW',
    ('C', 'Y'): 'OPPONENT',
    ('A', 'Z'): 'OPPONENT',
    ('B', 'Z'): 'ME',
    ('C', 'Z'): 'DRAW',
}
```

```python
total_score = 0
for l in lines:
    opp, me = l.split(' ')
    total_score += hand_values[me]
    match win_map[(opp, me)]:
        case 'DRAW':
            total_score += 3
        case 'ME':
            total_score += 6
total_score
```
