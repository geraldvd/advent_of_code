# Day 1, Exercise a

```python
with open("input_1.txt", "r") as fh:
    lines = fh.readlines()
# Remove \n characters
lines = [l.strip() for l in lines]
```

```python
elf_rations = []
cur_elf_ration = 0
for l in lines:
    if l == '':
        elf_rations.append(cur_elf_ration)
        cur_elf_ration = 0
    else:
        cur_elf_ration += int(l)
max(elf_rations)
```
